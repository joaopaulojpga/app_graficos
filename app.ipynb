```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
from io import BytesIO

# ConfiguraÃ§Ã£o inicial da pÃ¡gina
st.set_page_config(page_title="Gerador de GrÃ¡ficos de OcorrÃªncias", layout="wide")

st.title("ğŸ“Š Gerador de GrÃ¡ficos Semanais por Unidade")
st.write("Envie sua planilha e gere visualizaÃ§Ãµes agrupadas por semana, unidade e tipo de ocorrÃªncia.")

# Upload do arquivo
uploaded_file = st.file_uploader("Anexe a planilha (.xlsx)", type=["xlsx"])

if uploaded_file:
    # Leitura da planilha
    df = pd.read_excel(uploaded_file)

    # Renomear colunas para facilitar (caso nÃ£o tenham cabeÃ§alho exato)
    df.columns = df.columns.str.strip()

    # ValidaÃ§Ã£o das colunas necessÃ¡rias
    required_cols = ["Data de Abertura", "Unidade", "Tipo", "Status"]
    if not all(col in df.columns for col in required_cols):
        st.error(f"A planilha precisa conter as colunas: {', '.join(required_cols)}")
    else:
        # ConversÃ£o da coluna de data
        df["Data de Abertura"] = pd.to_datetime(df["Data de Abertura"], errors='coerce')
        df = df.dropna(subset=["Data de Abertura"])

        # CriaÃ§Ã£o da coluna de semana
        df["Semana"] = df["Data de Abertura"].dt.strftime("%Y-%U")

        # Agrupamento dos dados
        grouped = df.groupby(["Semana", "Unidade", "Tipo"]).size().reset_index(name="Contagem")

        # ConfiguraÃ§Ã£o do estilo ggplot
        plt.style.use('ggplot')

        unidades = grouped["Unidade"].unique()
        st.write(f"### GrÃ¡ficos Gerados ({len(unidades)} unidades encontradas)")

        for unidade in unidades:
            subset = grouped[grouped["Unidade"] == unidade]
            pivot = subset.pivot(index="Semana", columns="Tipo", values="Contagem").fillna(0)

            fig, ax = plt.subplots(figsize=(8, 4))
            pivot.plot(kind="barh", stacked=True, ax=ax)
            ax.set_title(f"{unidade} â€” OcorrÃªncias Semanais")
            ax.set_xlabel("Quantidade de OcorrÃªncias")
            ax.set_ylabel("Semana")
            st.pyplot(fig)
```
